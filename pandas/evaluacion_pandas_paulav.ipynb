# Preparaci√≥n de los datos:

- Los nombres de las columnas empiezan con may√∫sculas, poned todos los nombres de las columnas en min√∫sculas.

```python
def homogenizar_columnas (df):
    columnas_min = {col:col.lower() for col in df_datos.columns}
    df_datos.rename(columns = columnas_min , inplace = True)
```

- Hay algunas canciones cuyo t√≠tulo est√° en may√∫sulas. Crea una funci√≥n para que todos los t√≠tulos est√©n en min√∫scula. Haced lo mismo para las columnas de "track" y "album".

```python
def minuscula (df):

    df_datos["title"] = df_datos["title"].str.lower()
    df_datos["track"] = df_datos["track"].str.lower()
    df_datos["album"] = df_datos["album"].str.lower()

#Convertimos a string para poder hacerle el .lower(), 
# si no lo entiende com Serie y no permite ejecutar la funci√≥n.
```

```python
minuscula(df_datos)
```

- Algunos de los artistas tienen s√≠mbolos raros, en concreto el s√≠mbolo $. Cread una funci√≥n que nos elimine este s√≠mbolo de los nombres de los artistas.

```python
def simbolos (df):
    df_datos["artist"] = df_datos["artist"].str.replace("$","", regex=False)

#/tmp/ipykernel_21236/2856884554.py:2: FutureWarning: The default value of regex will change from True to False in a future version.
#In addition, single character regular expressions will *not* be treated as literal strings when regex=True.

#Al salirme este error y buscarlo en internet di con que la soluci√≥n era a√±adir regex=False,
#as√≠ se asegura que el simbolo "$" se trata como una cadena literal y no como un car√°cter especial.
```

Se quiere estudiar si existe una relaci√≥n entre la cantidad de visitas que reciben los videos de m√∫sica de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: 

- ¬øSon aquellos artistas que tienen m√°s visitas los que m√°s likes tienen? 

Para solucionar este ejercicio deber√©is:

- Agrupar por artista y calcular la media de visitas y de likes.

```python
df_visitas_likes = df_datos.groupby("artist")[["views", "likes"]].mean().sort_values(by= "views", ascending= False)
df_visitas_likes
```

- Quedaros con los 10 artistas que m√°s visitas han tenido.

```python
df_visitas = df_datos.groupby("artist")["views"].mean().reset_index().sort_values(by= "views", ascending= False)
df_visitas = df_visitas.head(10)
df_visitas
```

- Quedaros con los 10 artistas que m√°s likes han tenido.

```python
df_likes = df_datos.groupby("artist")["likes"].mean().reset_index().sort_values(by= "likes", ascending= False)
df_likes = df_likes.head(10)
df_likes
```

- Haced un subplot, donde en uno tengamos los 10 artistas con m√°s likes y en el otro los 10 artistas con mas views.

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(20, 10))

sns.barplot(x = df_likes["artist"], 
             y = df_likes["likes"], 
             color = "#DDA0DD",
             ax = axes[0])

sns.barplot(x = df_visitas["artist"],
             y = df_visitas["views"],
             color = "#7E1E9C",
             ax = axes[1])

axes[0].set_xlabel("Artistas")
axes[0].set_ylabel("Likes")
axes[1].set_xlabel("Artistas")
axes[1].set_ylabel("Visitas")

axes[0].set_title("TOP 10 ARTISTAS CON MAS LIKES", loc="center")
axes[1].set_title("TOP 10 ARTISTAS CON MAS VISITAS", loc="center")

plt.show()
```

Ahora queremos analizar la colaboraci√≥n entre artistas en la industria musical, se tienen datos de las colaboraciones realizadas en los √∫ltimos a√±os. Se desea conocer cu√°les son los artistas que han colaborado en mayor cantidad de canciones y generar un ranking con los diez primeros. 

- ¬øQu√© artistas con los que m√°s colaboraciones tienen? 

Para esto tendr√©is que:

- Filtrad los datos para quedaros solo con aquellos datos donde en la columna "track" tengamos "feat".

```python
df_feats = df_datos[df_datos["track"].str.contains("feat", regex= True)]
df_feats
```

- Contad cuantas canciones tiene cada artista utilizando el dataframe generado en el paso anterior.

```python
feats = df_feats.groupby('artist')['track'].count().reset_index().sort_values(by = "track", ascending= False)
feats
```

- Mostrad los resultados en una gr√°fica, seleccionado los 10 artistas con m√°s colaboraciones.

```python
df_mas_feats = feats.head(10)
df_mas_feats
```

```python
plt.figure(figsize =(20,10))
sns.barplot(x = df_mas_feats["track"], 
             y = df_mas_feats["artist"], 
             color = "#C79FEF")

plt.xlabel("Track")
plt.ylabel("Artista")

plt.title("TOP 10 ARTISTAS CON MAS Feats", loc="center")

plt.show()
```
