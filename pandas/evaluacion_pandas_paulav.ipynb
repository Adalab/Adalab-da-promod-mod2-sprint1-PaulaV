# Preparación de los datos:

- Los nombres de las columnas empiezan con mayúsculas, poned todos los nombres de las columnas en minúsculas.

```python
def homogenizar_columnas (df):
    columnas_min = {col:col.lower() for col in df_datos.columns}
    df_datos.rename(columns = columnas_min , inplace = True)
```

```python
homogenizar_columnas(df_datos)
```

- Haciendo el análisis exploratorio nos deberíamos haber dado cuenta de que algunas de las variables (danceability, energy, key, loudness, speechiness, acousticness, instrumentalness) no son del tipo que deberían. Esto es debido a que los decimales están establecidos como comas y no con puntos. Cread una función que nos permita cambiar esas comas por puntos para que los datos tengan el tipo correcto.

```python
df_datos.columns
#Al principio iba a hacer la función para todas columnas, 
# pero me di cuenta que podría afectar a otras variables con el tipo correcto,
# así que hice un .columns para iterar por las columnas que interesa cambiar. 
```

```python
def coma_punto (df):
    for i in df_datos.columns[6:13]:
        df_datos[i] = df_datos[i].apply(lambda x: float(x.replace(",",".")) if type(x) == str else x)


#Usamos el metodo apply lambda para poder aplicarlo a todos los valores especificados.   
```

```python
coma_punto(df_datos)
```

- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album".

```python
def minuscula (df):

    df_datos["title"] = df_datos["title"].str.lower()
    df_datos["track"] = df_datos["track"].str.lower()
    df_datos["album"] = df_datos["album"].str.lower()

#Convertimos a string para poder hacerle el .lower(), 
# si no lo entiende com Serie y no permite ejecutar la función.
```

```python
minuscula(df_datos)
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
def simbolos (df):
    df_datos["artist"] = df_datos["artist"].str.replace("$","", regex=False)

#/tmp/ipykernel_21236/2856884554.py:2: FutureWarning: The default value of regex will change from True to False in a future version.
#In addition, single character regular expressions will *not* be treated as literal strings when regex=True.

#Al salirme este error y buscarlo en internet di con que la solución era añadir regex=False,
#así se asegura que el simbolo "$" se trata como una cadena literal y no como un carácter especial.
```

Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: 

- ¿Son aquellos artistas que tienen más visitas los que más likes tienen? 

Para solucionar este ejercicio deberéis:

- Agrupar por artista y calcular la media de visitas y de likes.

```python
df_visitas_likes = df_datos.groupby("artist")[["views", "likes"]].mean().sort_values(by= "views", ascending= False)
df_visitas_likes
```

- Quedaros con los 10 artistas que más visitas han tenido.

```python
df_visitas = df_datos.groupby("artist")["views"].mean().reset_index().sort_values(by= "views", ascending= False)
df_visitas = df_visitas.head(10)
df_visitas
```

- Quedaros con los 10 artistas que más likes han tenido.

```python
df_likes = df_datos.groupby("artist")["likes"].mean().reset_index().sort_values(by= "likes", ascending= False)
df_likes = df_likes.head(10)
df_likes
```

- Haced un subplot, donde en uno tengamos los 10 artistas con más likes y en el otro los 10 artistas con mas views.

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize=(20, 10))

sns.barplot(x = df_likes["artist"], 
             y = df_likes["likes"], 
             color = "#DDA0DD",
             ax = axes[0])

sns.barplot(x = df_visitas["artist"],
             y = df_visitas["views"],
             color = "#7E1E9C",
             ax = axes[1])

axes[0].set_xlabel("Artistas")
axes[0].set_ylabel("Likes")
axes[1].set_xlabel("Artistas")
axes[1].set_ylabel("Visitas")

axes[0].set_title("TOP 10 ARTISTAS CON MAS LIKES", loc="center")
axes[1].set_title("TOP 10 ARTISTAS CON MAS VISITAS", loc="center")

plt.show()
```

Ahora queremos analizar la colaboración entre artistas en la industria musical, se tienen datos de las colaboraciones realizadas en los últimos años. Se desea conocer cuáles son los artistas que han colaborado en mayor cantidad de canciones y generar un ranking con los diez primeros. 

- ¿Qué artistas con los que más colaboraciones tienen? 

Para esto tendréis que:

- Filtrad los datos para quedaros solo con aquellos datos donde en la columna "track" tengamos "feat".

```python
df_feats = df_datos[df_datos["track"].str.contains("feat", regex= True)]
df_feats
```

- Contad cuantas canciones tiene cada artista utilizando el dataframe generado en el paso anterior.

```python
feats = df_feats.groupby('artist')['track'].count().reset_index().sort_values(by = "track", ascending= False)
feats
```

- Mostrad los resultados en una gráfica, seleccionado los 10 artistas con más colaboraciones.

```python
df_mas_feats = feats.head(10)
df_mas_feats
```

```python
plt.figure(figsize =(20,10))
sns.barplot(x = df_mas_feats["track"], 
             y = df_mas_feats["artist"], 
             color = "#C79FEF")

plt.xlabel("Track")
plt.ylabel("Artista")

plt.title("TOP 10 ARTISTAS CON MAS Feats", loc="center")

plt.show()
```
